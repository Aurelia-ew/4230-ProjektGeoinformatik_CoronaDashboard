```python
import altair as alt
import pandas as pd
import numpy as np

# Datensatz laden
source_file = "Corona_Zahlen_Kantone/COVID19_Fallzahlen_Kanton_BE_total.csv"
data_source = pd.DataFrame(pd.read_csv(source_file))
data_source
```

```python
data = data_source[["date", "abbreviation_canton_and_fl", "ncumul_conf", "current_hosp", "ncumul_deceased"]]
data["month"] = pd.to_datetime(data["date"]).dt.month
data
```

```python
ticks = ["2020-02-25", "2020-02-28", "2020-03-16", "2020-04-27", "2020-05-11", "2020-07-06", "2020-10-19","2020-12-23", "2021-01-18", "2021-03-01", "2021-09-13", "2021-12-20", "2022-02-17"]

lines = alt.Chart(data).mark_line(color="green").encode(
    x=alt.X('date:T',axis=alt.Axis(title="Datum", format="%d.%m.%y", values=ticks, labelAngle=-50, labelOverlap=False, labelFontSize=10)),
    y=alt.Y('current_hosp', title="Hospitalisierungen", scale=alt.Scale(domain=[0, 450]))
).properties(
    width="container",
    height=400
)
lines
```

```python
lines.to_json()
with open("Hosp_BE.json", "w") as f:
    f.write(lines.to_json())
```
